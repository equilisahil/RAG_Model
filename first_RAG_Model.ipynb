```python
from google.colab import userdata
import os
os.environ['GROQ_API_KEY'] = userdata.get('first_groq_key')
```

```python
from langchain_community.document_loaders import WebBaseLoader
loader = WebBaseLoader(web_paths=["https://www.geeksforgeeks.org/advantages-and-disadvantages-of-blockchain/"])
data = loader.load()
print(data)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 1200,
    chunk_overlap  = 800
)
texts = text_splitter.split_documents(data)
print(texts)
```

```python
print(len(texts))
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
# Step 1: Use a local embedding model (like from sentence-transformers)
embedding_model = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

# Step 2: Embed and store in Chroma
from langchain.vectorstores import Chroma

vectorstore = Chroma.from_documents(documents=texts, embedding= embedding_model)
```

```python
print(vectorstore._collection.count())
```

```python
print(vectorstore._collection.get())
```

```python
print("/nCollection-1" , vectorstore._collection.get(ids=['1d51d614-0c12-46fa-9a81-437d2be50df6']))
```

```python
# RAG pipeline Retrieval
retriever = vectorstore.as_retriever()
```

```python
# Augumentation
from langchain import hub
prompt = hub.pull("rlm/rag-prompt")
```

```python
# set up LLM
# from langchain.llms import HuggingFacePipeline
# from langchain.chains import RetrievalQA
from langchain_groq import ChatGroq
llm = ChatGroq(temperature=0,
              model_name="llama3-8b-8192",
              api_key=userdata.get("first_groq_key"))
```

```python
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
```

```python
def format_data(data):
  return "/n".join(datas.page_content for datas in data)
```

```python
# RAG chain code
rag_chain = ({"context": retriever | format_data, "question": RunnablePassthrough()}) | prompt | llm | StrOutputParser()
```

```python
rag_chain.invoke("What are the advantages of blockchain?")
```

```python
rag_chain.invoke("What is blockchain?")
```

```python
rag_chain.invoke("give me five important points of blockchain?")
```

```python
from langchain_core.runnables import RunnableLambda
```

```python
def print_prompt(prompt_text):
  print("prompt - ", prompt_text)
  return prompt_text
```

```python

rag_chain_with_print = ({"context": retriever | format_data, "question": RunnablePassthrough()}) | prompt | RunnableLambda(print_prompt) | llm | StrOutputParser()
```

```python
rag_chain_with_print.invoke("What are the advantages of blockchain?")
```
